```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd

# 1. WebDriver 설정
driver = webdriver.Chrome()  # chromedriver의 경로를 지정해야 할 수 있습니다
url = "https://exhibitors.ces.tech/8_0/#/searchtype/category/search/999999999/show/cat-category%7C240"  # 크롤링할 대상 URL로 교체하세요
driver.get(url)

# 2. 페이지 로딩 대기
time.sleep(3)

# 3. 스크롤 및 더 많은 콘텐츠 로드
last_height = driver.execute_script("return document.body.scrollHeight")  # 초기 스크롤 높이 가져오기
scraped_data = []  # 데이터를 저장할 리스트

while True:
    # 페이지의 맨 아래로 스크롤
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    
    # 새로운 콘텐츠가 로드되도록 대기
    time.sleep(2)  # 페이지 로딩 시간에 따라 대기 시간을 조정하세요
    
    # 새로운 스크롤 높이 가져오기
    new_height = driver.execute_script("return document.body.scrollHeight")
    
    # 스크롤 높이에 변화가 없으면 반복 종료
    if new_height == last_height:
        print("더 이상 로드할 콘텐츠가 없습니다.")
        break
    last_height = new_height

# 4. 스크롤 후 데이터 추출
data_elements = driver.find_elements(By.CLASS_NAME, 'card-Title break-word f2 mb1 mt0')  # 실제 클래스 이름으로 교체하세요
for element in data_elements:
    scraped_data.append(element.text)  # 텍스트 데이터를 리스트에 추가

# 5. 브라우저 닫기
driver.quit()

# 6. 데이터 저장: Pandas DataFrame으로 변환 및 Excel 저장
df = pd.DataFrame(scraped_data, columns=["Content"])  # 데이터 컬럼 이름 수정 가능
excel_file_name = "scraped_data.xlsx"  # 저장할 파일 이름
df.to_excel(excel_file_name, index=False, encoding="utf-8")
print(f"데이터가 '{excel_file_name}'로 저장되었습니다.")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd  # Import pandas for saving data to Excel

# 1. WebDriver 설정
driver = webdriver.Chrome()  # chromedriver 경로를 지정해야 할 수 있습니다
url = "https://exhibitors.ces.tech/8_0/#/searchtype/category/search/999999999/show/cat-category%7C240"  # CES 2025 URL
driver.get(url)

# 2. 페이지 로딩 대기
time.sleep(15)

# 3. 스크롤 및 더 많은 콘텐츠 로드
last_height = driver.execute_script("return document.body.scrollHeight")  # 초기 스크롤 높이
scraped_data = []  # 데이터를 저장할 리스트

while True:
    # 페이지의 맨 아래로 스크롤
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    
    # 새로운 콘텐츠가 로드되도록 대기
    time.sleep(2)  # 페이지 로딩 시간에 따라 대기 시간 조정
    
    # 새로운 스크롤 높이 확인
    new_height = driver.execute_script("return document.body.scrollHeight")
    
    # 스크롤 높이에 변화가 없으면 루프 종료
    if new_height == last_height:
        print("더 이상 로드할 콘텐츠가 없습니다.")
        break
    last_height = new_height

# 4. 데이터 추출: p 태그 클래스가 'card-Desc f5 lh-copy ma0 muted'인 요소 내의 a 태그 찾기
p_elements = driver.find_elements(By.CLASS_NAME, 'card-Desc f5 lh-copy ma0 muted')  # p 요소를 찾음
for p in p_elements:
    # p 내의 a 태그 찾기
    a_tag = p.find_element(By.TAG_NAME, 'a')  # a 태그 찾기
    
    # a 태그 내의 href 속성값 추출
    href_value = a_tag.get_attribute('href')
    
    # 찾은 href 값에 대한 추가적인 클래스가 'bb-0'인 요소 내의 데이터를 추출
    bb0_element = a_tag.find_element(By.CLASS_NAME, 'bb-0')  # 'bb-0' 클래스를 가진 요소 찾기
    if bb0_element:
        scraped_data.append({"Link": href_value, "BB0 Content": bb0_element.text})  # 링크와 내용 저장

# 5. 브라우저 닫기
driver.quit()

# 6. 데이터 저장: Pandas DataFrame으로 변환 및 Excel 저장
df = pd.DataFrame(scraped_data)  # DataFrame으로 변환
output_file = "scraped_data.xlsx"  # 저장할 파일 이름

# Excel 파일로 저장 (encoding 제거)
df.to_excel(output_file, index=False, engine="openpyxl")
print(f"데이터가 엑셀 파일로 저장되었습니다: {output_file}")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time
import logging
from typing import List, Dict
import os

logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s',
    handlers=[
        logging.FileHandler('scraper.log'),
        logging.StreamHandler()
    ]
)

class CESExhibitorScraper:
    def __init__(self):
        self.chrome_options = Options()
        self.chrome_options.add_argument("--headless")
        self.chrome_options.add_argument("--disable-gpu")
        self.chrome_options.add_argument("--no-sandbox")
        self.chrome_options.add_argument("--disable-dev-shm-usage")
        self.chrome_options.add_argument("--window-size=1920,1080")
        
        # User-Agent 추가
        self.chrome_options.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36")
        
        self.service = Service(ChromeDriverManager().install())
        self.driver = None
        
    def __enter__(self):
        self.driver = webdriver.Chrome(service=self.service, options=self.chrome_options)
        self.wait = WebDriverWait(self.driver, 30)  # 대기 시간을 30초로 증가
        return self
        
    def __exit__(self, exc_type, exc_val, exc_tb):
        if self.driver:
            self.driver.quit()
            
    def wait_for_page_load(self):
        """페이지가 완전히 로드될 때까지 대기"""
        try:
            self.wait.until(lambda driver: driver.execute_script('return document.readyState') == 'complete')
            time.sleep(5)  # 추가 대기 시간
        except TimeoutException:
            logging.warning("페이지 로드 완료 대기 시간 초과")
            
    def scroll_page(self, timeout: int = 60) -> None:  # 타임아웃 증가
        """페이지를 천천히 스크롤하며 모든 콘텐츠 로드"""
        start_time = time.time()
        last_height = self.driver.execute_script("return document.body.scrollHeight")
        
        while True:
            # 천천히 스크롤
            for i in range(10):
                self.driver.execute_script(f"window.scrollTo(0, {last_height * (i/10)});")
                time.sleep(0.5)
            
            time.sleep(3)  # 로딩 대기 시간 증가
            
            new_height = self.driver.execute_script("return document.body.scrollHeight")
            if new_height == last_height or time.time() - start_time > timeout:
                break
            last_height = new_height
            
    def get_exhibitor_links(self, url: str) -> List[str]:
        """메인 페이지에서 모든 전시업체 링크 가져오기"""
        try:
            self.driver.get(url)
            logging.info("메인 페이지 접속: %s", url)
            
            # 페이지 로드 대기
            self.wait_for_page_load()
            
            # Accept 버튼이 있는지 확인하고 클릭
            try:
                accept_button = self.driver.find_element(By.CSS_SELECTOR, "button[aria-label='Accept']")
                accept_button.click()
                time.sleep(2)
            except NoSuchElementException:
                pass
            
            # 요소가 나타날 때까지 대기
            try:
                self.wait.until(EC.presence_of_element_located(
                    (By.CSS_SELECTOR, "h3 card-Title break-word dib f5 ma0 a")
                ))
            except TimeoutException:
                logging.error("전시업체 카드를 찾을 수 없습니다.")
                return []
            
            self.scroll_page()
            
            # 여러 번 시도하여 링크 수집
            max_attempts = 3
            for attempt in range(max_attempts):
                elements = self.driver.find_elements(
                    By.CSS_SELECTOR, 
                    "h3 card-Title break-word dib f5 ma0 a"
                )
                links = [element.get_attribute("href") for element in elements]
                
                if links:
                    logging.info("전시업체 링크 %d개 발견 (시도 %d)", len(links), attempt + 1)
                    return links
                    
                if attempt < max_attempts - 1:
                    logging.warning("링크를 찾지 못했습니다. 재시도 중... (%d/%d)", attempt + 1, max_attempts)
                    time.sleep(5)
                    self.scroll_page()
            
            logging.error("최대 시도 횟수 초과")
            return []
            
        except Exception as e:
            logging.error("전시업체 링크 수집 중 오류: %s", str(e))
            return []
            
    def get_exhibitor_data(self, links: List[str]) -> List[Dict]:
        """각 전시업체 페이지에서 데이터 수집"""
        data = []
        
        for i, link in enumerate(links, 1):
            try:
                self.driver.get(link)
                self.wait_for_page_load()
                logging.info("[%d/%d] 처리 중: %s", i, len(links), link)
                
                # 주소 요소 대기
                try:
                    address_element = self.wait.until(EC.presence_of_element_located(
                        (By.CSS_SELECTOR, "showcase-address tc")
                    ))
                except TimeoutException:
                    logging.warning("주소를 찾을 수 없습니다: %s", link)
                    continue
                
                # 추가 정보 수집
                company_name = self.safe_get_text(".company-name")
                booth_number = self.safe_get_text(".booth-number")
                description = self.safe_get_text(".company-description")
                
                data.append({
                    "Link": link,
                    "Company": company_name,
                    "Address": address_element.text,
                    "Booth": booth_number,
                    "Description": description
                })
                
                time.sleep(2)  # 요청 간 지연 시간 증가
                
            except Exception as e:
                logging.error("링크 처리 중 오류 %s: %s", link, str(e))
                
        return data
        
    def safe_get_text(self, selector: str) -> str:
        """요소가 존재하는 경우 안전하게 텍스트 가져오기"""
        try:
            element = self.driver.find_element(By.CSS_SELECTOR, selector)
            return element.text.strip()
        except NoSuchElementException:
            return ""
            
    def save_to_excel(self, data: List[Dict], filename: str) -> None:
        """데이터를 Excel 파일로 저장"""
        try:
            df = pd.DataFrame(data)
            df.to_excel(filename, index=False, engine="openpyxl")
            logging.info("데이터가 %s 파일에 성공적으로 저장됨", filename)
        except Exception as e:
            logging.error("Excel 저장 중 오류: %s", str(e))

def main():
    url = "https://exhibitors.ces.tech/8_0/#/searchtype/category/search/999999999/show/cat-category%7C215"
    output_file = "ces_exhibitors_data.xlsx"
    
    try:
        with CESExhibitorScraper() as scraper:
            links = scraper.get_exhibitor_links(url)
            if not links:
                logging.error("전시업체 링크를 찾을 수 없습니다. 종료합니다.")
                return
                
            data = scraper.get_exhibitor_data(links)
            if data:
                scraper.save_to_excel(data, output_file)
            else:
                logging.error("수집된 데이터가 없습니다. 종료합니다.")
                
    except Exception as e:
        logging.error("치명적 오류: %s", str(e))

if __name__ == "__main__":
    main()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd

# 1. WebDriver 설정
options = Options()
options.add_argument("--start-maximized")
options.add_argument("--disable-blink-features=AutomationControlled")
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36")

driver = webdriver.Chrome(options=options)
url = "https://exhibitors.ces.tech/8_0/explore/exhibitor-gallery.cfm?featured=false"
driver.get(url)

# 2. 페이지 로딩 대기
time.sleep(10)

# 3. 'is-selected toggle-grid btn-tertiary' 버튼 클릭
try:
    button = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.CSS_SELECTOR, "span.is-selected.toggle-grid.btn-tertiary"))
    )
    button.click()
    print("✅ 버튼 클릭 완료!")
except Exception as e:
    print(f"🚨 버튼 클릭 실패: {e}")

# 4. 페이지 로딩 대기
time.sleep(10)

# 5. 페이지 끝까지 스크롤
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(10)  # 페이지 로딩 대기

    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        print("📄 모든 콘텐츠 로드 완료!")
        break
    last_height = new_height

# 6. 데이터 추출
scraped_data = []

try:
    # 전시명(Exhibitor) 탐색
    exhibitors = WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'h3.card-Title.break-word.dib.f5.ma0'))
    )
    # 설명(Summary) 탐색
    descriptions = driver.find_elements(By.CSS_SELECTOR, 'td.is-hidden_small.card-Desc.f6 span')

    # 전시명 & 설명 내용 매칭
    for exhibitor, description in zip(exhibitors, descriptions):
        exhibitor_name = exhibitor.text.strip() if exhibitor.text else "N/A"
        summary_text = description.text.strip() if description.text else "N/A"

        scraped_data.append({
            "Exhibitor": exhibitor_name,
            "Summary": summary_text
        })

except Exception as e:
    print(f"🚨 데이터 탐색 중 오류 발생: {e}")

# 7. 브라우저 종료
driver.quit()

# 8. DataFrame 생성 및 Excel 저장
df = pd.DataFrame(scraped_data)
output_file = "CES_Exhibitors_Summary_with_button.xlsx"
df.to_excel(output_file, index=False, engine="openpyxl")

print(f"💾 데이터가 '{output_file}'에 저장되었습니다.")
```
