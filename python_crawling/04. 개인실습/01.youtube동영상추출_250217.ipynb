```python
import os
import yt_dlp
import time
from urllib.parse import urlparse

# 1️⃣ ffmpeg 경로 설정 (Windows 환경)
os.environ["PATH"] += os.pathsep + r"C:\ffmpeg\bin"

# 2️⃣ 유튜브 동영상 형식 확인
def list_formats(video_url):
    try:
        with yt_dlp.YoutubeDL({'listformats': True}) as ydl:
            info = ydl.extract_info(video_url, download=False)
            print("\n[INFO] 사용 가능한 형식:")
            for fmt in info['formats']:
                print(f"ID: {fmt['format_id']}, Resolution: {fmt['resolution']}, Size: {fmt.get('filesize', 'Unknown')} bytes")
    except Exception as e:
        print(f"[ERROR] 형식 목록 가져오는 중 오류 발생: {e}")

# 3️⃣ 유튜브 동영상 다운로드 함수
def download_video(video_url, output_filename):
    ydl_opts = {
        'format': 'bestvideo[ext=mp4]+bestaudio[ext=m4a]/best',
        'outtmpl': f'downloads/{output_filename}.%(ext)s',
        'ffmpeg_location': r'C:\ffmpeg\bin',
        'writesubtitles': True,
        'writeautomaticsub': True,
        'subtitleslangs': ['en', 'ko'],
        'subtitlesformat': 'vtt',
        'postprocessors': [{
            'key': 'FFmpegVideoConvertor',
            'preferedformat': 'mp4',
        }],
    }
    try:
        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            ydl.download([video_url])
            print(f"[INFO] 다운로드 완료: {output_filename}")
    except Exception as e:
        print(f"[ERROR] 다운로드 실패: {video_url}, 오류: {e}")
        return False
    return True

# 4️⃣ 자막 파일 저장 경로 확인
def print_subtitle_location(output_filename):
    subtitle_files = [
        f"downloads/{output_filename}.en.vtt",
        f"downloads/{output_filename}.ko.vtt"
    ]
    for file in subtitle_files:
        if os.path.exists(file):
            print(f"[INFO] 자막 파일 저장 위치: {os.path.abspath(file)}")
        else:
            print(f"[WARNING] 자막 파일이 없습니다: {file}")

# 5️⃣ 실행
if __name__ == "__main__":
    os.makedirs("downloads", exist_ok=True)
    video_url = "https://www.youtube.com/watch?v=rCcxXxoerjg"#https://www.youtube.com/watch?v=XVo--k1JCrA #지멘스 소니영상
    list_formats(video_url)
    output_filename = "my_video"
    if download_video(video_url, output_filename):
        print_subtitle_location(output_filename)
```
