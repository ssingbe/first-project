```python
import openai
import os

# Set the API key directly
openai.api_key = "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"  # Replace with your actual API key

# Define a function to read a script file
def read_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        return file.read()

# Define a function to translate content using ChatGPT
def translate_content(content, model="text-davinci-003", prompt_format=None):
    prompt = f"Summarize the contents of the attached file as follows.\n\n" \
             "Think about it step by step and translate it into Korean.\n\n" \
             "#work\nGenerate conference summary and presentation materials\n\n" \
             "#Conference_Information\n" \
             "Title: [Enter conference title]\n" \
             "Topic: [Enter main topic]\n" \
             "Date: [Enter date]\n" \
             "Organizer: [Enter Host Organization]\n\n" \
             "#Summary_Guidelines\n" \
             "- Summary of key speakers and their key messages\n" \
             "- Identify the 3-5 most important topics or trends\n" \
             "- Innovative ideas or technology highlights\n" \
             "- Analysis of potential impact on the industry\n\n" \
             "#presentation_data_structure\n" \
             "## Slide 1: Conference Overview\n" \
             "## Slides 2-4: Key topics and trends (1 slide per topic)\n" \
             "## Slide 5: Innovative Ideas and Technologies\n" \
             "## Slide 6: Industry Impact Analysis\n" \
             "## Slide 7: Conclusion and Key Takeaways\n\n" \
             "#presentation_style\n" \
             "- Use concise and clear sentences\n" \
             "- Limit to 3-5 key points per slide\n" \
             "- Include relevant statistics or data (if available)\n" \
             "- Suggestions for visual elements (charts, graphs, icons, etc.)\n\n" \
             "#Target_Audience\n" \
             "[Enter target audience characteristics, e.g. executives, technical experts, general employees, etc.]\n\n" \
             "#Presentation_Time\n" \
             "[Enter expected presentation time, e.g. 15 minutes]\n\n" \
             "#Additional_Requests\n" \
             "- Suggest 2-3 questions or discussion topics for audience interaction\n" \
             "- Create a reference list for additional information or resources.\n\n" \
             f"{content}"

    try:
        response = openai.Completion.create(
            model=model,
            prompt=prompt,
            max_tokens=1500,
            temperature=0.7
        )
        return response.choices[0].text.strip()
    except Exception as e:
        print(f"Error translating content: {e}")
        return None

# Define a function to save the translated content
def save_translated_file(content, output_folder, original_file_name):
    os.makedirs(output_folder, exist_ok=True)
    output_file_path = os.path.join(output_folder, original_file_name)
    with open(output_file_path, 'w', encoding='utf-8') as file:
        file.write(content)

# Define the main function to process all files in a folder
def process_folder(input_folder, output_folder, model="text-davinci-003", prompt_format=None):
    for file_name in os.listdir(input_folder):
        if file_name.endswith('.txt'):  # Process only .txt files
            file_path = os.path.join(input_folder, file_name)
            content = read_file(file_path)
            translated_content = translate_content(content, model=model, prompt_format=prompt_format)
            if translated_content:
                save_translated_file(translated_content, output_folder, file_name)

# Specify folders
input_folder = r"C:/Temp/input_scripts"  # Folder containing scripts
output_folder = "translated_scripts"  # Folder to save translations

# Run the program
if __name__ == "__main__":
    process_folder(input_folder, output_folder)
```
