```python
# Importing libraries
import pandas as pd
```

```python
# 투입량 Loading
INPUT_WT = pd.read_csv('input_hangul.csv', encoding='cp949')
INPUT_WT.head()
```

```python
# 중강종 파생변수 생성코드
INPUT_WT['중강종'] = INPUT_WT['강종그룹'].str[0:2]
# STEEL_GROUP_L는 Steel Grade 컬럼 앞에 배치
position = INPUT_WT.columns.get_loc('강종그룹')
INPUT_WT.insert(position, '중강종', INPUT_WT.pop('중강종'))

# 대강종 파생변수 생성코드
INPUT_WT['대강종'] = INPUT_WT['강종그룹'].str[0]
# 대강종은 중강종 컬럼 앞에 배치
position = INPUT_WT.columns.get_loc('중강종')
INPUT_WT.insert(position, '대강종', INPUT_WT.pop('대강종'))

INPUT_WT.head()
#print(INPUT_WT['대강종'])
```

```python
# Year Month Date' 열을 datetime 형식으로 변환
INPUT_WT['최종압연일자'] = pd.to_datetime(INPUT_WT['최종압연일자'])

#압연월 구분
INPUT_WT['압연월'] = INPUT_WT['최종압연일자'].dt.month
#INPUT_WT['압연월'] = INPUT_WT['압연월'].fillna("", inplace=True)  # Fill missing values with "" (or any other appropriate value)

# 압연월은 년월 컬럼 앞에 배치
position = INPUT_WT.columns.get_loc('년월')
INPUT_WT.insert(position, '압연월', INPUT_WT.pop('압연월'))

#압연년 구분
INPUT_WT['압연년'] = INPUT_WT['최종압연일자'].dt.year
#INPUT_WT['압연년'] = INPUT_WT['압연년'].fillna("", inplace=True)  # Fill missing values with "" (or any other appropriate value)

# 압연년은 압연월 컬럼 앞에 배치
position = INPUT_WT.columns.get_loc('압연월')
INPUT_WT.insert(position, '압연년', INPUT_WT.pop('압연년'))

INPUT_WT.head()
```

```python
# Check if '압연월' is empty and replace '압연분기' with a blank space
mask = INPUT_WT['압연월'] == ""
INPUT_WT.loc[mask, '압연분기'] = ""

# Calculate '압연분기' for rows with non-empty '압연월'
mask = INPUT_WT['압연월'] != ""
INPUT_WT.loc[mask, '압연분기'] = INPUT_WT.loc[mask, '압연월'].apply(lambda month: '1Q' if month <= 3 else '2Q' if month <= 6 else '3Q' if month <= 9 else '4Q')

# 압연분기는 압연월 컬럼 앞에 배치
position = INPUT_WT.columns.get_loc('압연월')
INPUT_WT.insert(position, '압연분기', INPUT_WT.pop('압연분기'))

INPUT_WT.head()
#print(INPUT_WT['압연분기'])
```

```python
# 1,433처럼 중량에 콤마가 있는 Object는 콤마를 제거 후 숫자로 변환해야 함.
INPUT_WT['투입중량'] = INPUT_WT['투입중량'].str.replace(',', '').astype(int)
INPUT_WT['생산중량'] = INPUT_WT['생산중량'].str.replace(',', '').astype(int)
INPUT_WT.head()
```

```python
# 엑셀 파일로 저장
# pip install openpyxl 사전 설치 필요
output_path = 'pre_input_data.xlsx'  # 저장할 파일 경로와 이름
INPUT_WT.to_excel(output_path, index=False)  # index=False로 설정하여 인덱스를 엑셀에 포함하지 않음
```
