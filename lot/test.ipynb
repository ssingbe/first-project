```python
import pandas as pd

# 예제 데이터 생성
data = {
    '년도': [2024, 2024, 2024, 2023, 2023],
    '제품형태': ['형태A', '형태B', '형태A', '형태A', '형태B'],
    '투입중량': [1000, 1500, 1200, 800, 1000],
    '생산중량': [800, 1200, 950, 600, 800]
}

# DataFrame 생성
df = pd.DataFrame(data)

# 제품 형태 및 년도별 집계
by_year_and_type = df.groupby(['제품형태', '년도']).agg({
    '투입중량': 'sum',
    '생산중량': 'sum'
}).reset_index()

# 수율 계산
by_year_and_type['수율'] = by_year_and_type['생산중량'] / by_year_and_type['투입중량']

# 제품 형태별 중간 합계 집계
by_type_total = df.groupby('제품형태').agg({
    '투입중량': 'sum',
    '생산중량': 'sum'
}).reset_index()

# 수율 계산
by_type_total['수율'] = by_type_total['생산중량'] / by_type_total['투입중량']

# 중간 합계에 총계 행 추가
by_type_total.loc['총계'] = by_type_total[['투입중량', '생산중량', '수율']].sum()
by_type_total.at['총계', '제품형태'] = '총계'

# 결과 출력
#print("년도 및 제품 형태별 집계:")
#print(by_year_and_type)
#print("\n제품 형태별 중간 합계:")
#print(by_type_total)
by_year_and_type.head()
by_type_total.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 데이터 생성 (임의 데이터 포함)
data = {
    '년월': [
        202105, 202105, 202105, 202105, 202106, 202106, 202112, 202201, 202202, 202203, 
        202207, 202212, 202212, 202301, 202301, 202306, 202307, 202311, 202312, 202401, 
        202402, 202402
    ],
    '생산': np.random.randint(800, 3000, size=22)  # 임의의 생산량 데이터
}

df = pd.DataFrame(data)

# X축 구간 설정 함수
def categorize_period(년월):
    if str(년월).startswith('2021'):
        return '21년'
    elif str(년월).startswith('2022'):
        return '22년'
    elif '202301' <= str(년월) <= '202306':
        return '23년 상반기'
    elif '202307' <= str(년월) <= '202312':
        return '23년 하반기'
    elif str(년월) == '202401':
        return '24년1월'
    elif str(년월) == '202402':
        return '24년2월'
    return '기타'

# 새로운 컬럼 추가 (구간)
df['기간'] = df['년월'].apply(categorize_period)

# 각 구간별 생산량의 평균을 계산
df_grouped = df.groupby('기간')['생산'].mean().reset_index()

# 그래프 그리기
plt.figure(figsize=(10, 6))
plt.bar(df_grouped['기간'], df_grouped['생산'], color='skyblue')
plt.xlabel('기간')
plt.ylabel('생산량')
plt.title('기간별 생산량')
plt.xticks(rotation=45)
plt.tight_layout()

# 그래프 보여주기
plt.show()
```

```python
# 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'  # Windows의 경우 '맑은 고딕'

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 데이터 생성 (년월 길이에 맞는 임의 데이터)
data = {
    '년월': [
        202101, 202102, 202103, 202104, 202105, 202106, 202107, 202108, 202109, 202110, 202111, 202112, 
        202201, 202202, 202203, 202204, 202205, 202206, 202207, 202208, 202209, 202210, 202211, 202212,
        202301, 202302, 202303, 202304, 202305, 202306, 202307, 202308, 202309, 202310, 202311, 202312,
        202401, 202402
    ],
    '생산': np.random.randint(800, 3000, size=38),  # 년월 데이터 길이(37)에 맞춘 생산량 데이터
    '투입': np.random.randint(1000, 3500, size=38)  # 임의의 투입 데이터
}

df = pd.DataFrame(data)

# Y값 계산 (생산/투입 * 100)
df['효율'] = (df['생산'] / df['투입']) * 100
df['효율'] = df['효율'].round(1)  # 소수점 1자리까지 반올림

# X축 구간 설정 함수 (분기 및 합쳐진 기간 포함)
def categorize_period(년월):
    year = str(년월)[:4]  # 연도
    month = int(str(년월)[4:])  # 월
    
    if year == '2021':
        if month <= 3:
            return '21년 1분기'
        elif 4 <= month <= 6:
            return '21년 2분기'
        elif 7 <= month <= 9:
            return '21년 3분기'
        else:
            return '21년 4분기'
    elif year == '2022':
        if month <= 3:
            return '22년 1분기'
        elif 4 <= month <= 6:
            return '22년 2분기'
        elif 7 <= month <= 9:
            return '22년 3분기'
        else:
            return '22년 4분기'
    elif year == '2023':
        if month <= 6:
            return '23년 상반기'
        else:
            return '23년 하반기'
    elif year == '2024':
        return '23년 하반기-24년 1~2월'  # 2024년 1, 2월을 2023년 하반기와 합침
    return '기타'

# 새로운 컬럼 추가 (구간)
df['기간'] = df['년월'].apply(categorize_period)

# 각 구간별 효율의 평균을 계산
df_grouped = df.groupby('기간')['효율'].mean().reset_index()

# 그래프 그리기
plt.figure(figsize=(12, 6))
bars = plt.bar(df_grouped['기간'], df_grouped['효율'], color='lightgreen')

# 막대 위에 값 표시
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:.1f}%', ha='center', va='bottom')

# 그래프 레이블 및 제목 설정
plt.xlabel('기간')
plt.ylabel('효율 (%)')
plt.title('분기 및 합쳐진 기간별 생산 효율')
plt.xticks(rotation=45)
plt.tight_layout()

# 그래프 보여주기
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 데이터 생성 (년월 및 형상 포함 임의 데이터)
data = {
    '년월': [
        202101, 202102, 202103, 202104, 202105, 202106, 202107, 202108, 202109, 202110, 202111, 202112, 
        202201, 202202, 202203, 202204, 202205, 202206, 202207, 202208, 202209, 202210, 202211, 202212,
        202301, 202302, 202303, 202304, 202305, 202306, 202307, 202308, 202309, 202310, 202311, 202312,
        202401, 202402
    ],
    '형상': ['A', 'B', 'C'] * 12 + ['A', 'B'],  # 형상: A, B, C를 반복하여 총 37개로 맞춤
    '생산': np.random.randint(800, 3000, size=38),  # 년월 데이터 길이(37)에 맞춘 생산량 데이터
    '투입': np.random.randint(1000, 3500, size=38)  # 임의의 투입 데이터
}

df = pd.DataFrame(data)

# Y값 계산 (생산/투입 * 100)
df['효율'] = (df['생산'] / df['투입']) * 100
df['효율'] = df['효율'].round(1)  # 소수점 1자리까지 반올림

# X축 구간 설정 함수 (분기 및 합쳐진 기간 포함)
def categorize_period(년월):
    year = str(년월)[:4]  # 연도
    month = int(str(년월)[4:])  # 월
    
    if year == '2021':
        if month <= 3:
            return '21년 1분기'
        elif 4 <= month <= 6:
            return '21년 2분기'
        elif 7 <= month <= 9:
            return '21년 3분기'
        else:
            return '21년 4분기'
    elif year == '2022':
        if month <= 3:
            return '22년 1분기'
        elif 4 <= month <= 6:
            return '22년 2분기'
        elif 7 <= month <= 9:
            return '22년 3분기'
        else:
            return '22년 4분기'
    elif year == '2023':
        if month <= 6:
            return '23년 상반기'
        else:
            return '23년 하반기'
    elif year == '2024':
        return '23년 하반기-24년 1~2월'  # 2024년 1, 2월을 2023년 하반기와 합침
    return '기타'

# 새로운 컬럼 추가 (구간)
df['기간'] = df['년월'].apply(categorize_period)

# 형상 및 기간별 평균 효율 계산
df_grouped = df.groupby(['기간', '형상'])['효율'].mean().unstack()

# 그래프 그리기
df_grouped.plot(kind='bar', figsize=(12, 6))

# 그래프 레이블 및 제목 설정
plt.xlabel('기간')
plt.ylabel('효율 (%)')
plt.title('형상별 분기 및 합쳐진 기간별 생산 효율')
plt.xticks(rotation=45)
plt.legend(title='형상')
plt.tight_layout()

# 그래프 보여주기
plt.show()
```
