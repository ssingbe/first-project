```python
#단변량 분석
#개별 변수 탐색
#숫자변수에 대한 히스토그램
df['외경'].plot.hist()
plt.show()
```

```python
#단변량 분석
#개별 변수 탐색
#범주형 변수에 대한 개수 도표:
sns.countplot(x='강종그룹', data=df)
plt.show()
```

```python
#이변량 분석
#두 변수 사이의 관계
#숫자 변수에 대한 산점도:
sns.scatterplot(x='외경', y='폭', data=df)
plt.show()
```

```python
#범주별 분포를 시각화하는 상자 그림:
df['발생중량'] = pd.to_numeric(df['발생중량'], errors='coerce')
sns.boxplot(x='강종그룹', y='발생중량', data=df)
plt.show()
```

```python
# Select only numeric columns for pair plotting
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Create a pair plot
sns.pairplot(numeric_df, diag_kind='kde')
plt.show()
```

```python
correlation_matrix = numeric_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```
