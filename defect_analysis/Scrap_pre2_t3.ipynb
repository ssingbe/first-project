```python
# Importing libraries
import pandas as pd
```

```python
# 불량량 Loading
Defect_WT = pd.read_csv('S300_scrap_output_hangul4.csv', encoding='cp949')
Defect_WT.head()
```

```python
# 중강종 파생변수 생성코드
Defect_WT['중강종'] = Defect_WT['강종그룹'].str[0:2]
# STEEL_GROUP_L는 Steel Grade 컬럼 앞에 배치
position = Defect_WT.columns.get_loc('강종그룹')
Defect_WT.insert(position, '중강종', Defect_WT.pop('중강종'))
Defect_WT.head()
#print(Defect_WT['중강종'])
```

```python
# 대강종 파생변수 생성코드
Defect_WT['대강종'] = Defect_WT['강종그룹'].str[0]
# 대강종은 중강종 컬럼 앞에 배치
position = Defect_WT.columns.get_loc('중강종')
Defect_WT.insert(position, '대강종', Defect_WT.pop('대강종'))
Defect_WT.head()
```

```python
#외경구분
Defect_WT['외경구분'] = Defect_WT['외경'].apply(lambda x: 'Ø15~25' if x <= 25.5 else 'Ø26~36')
position = Defect_WT.columns.get_loc('외경')
Defect_WT.insert(position, '외경구분', Defect_WT.pop('외경구분'))
Defect_WT.head()
#print(Defect_WT)
```

```python
# Year Month Date' 열을 datetime 형식으로 변환
Defect_WT['최종압연일자'] = pd.to_datetime(Defect_WT['최종압연일자'])

#압연월 구분
Defect_WT['압연월'] = Defect_WT['최종압연일자'].dt.month
#Defect_WT['압연월'] = Defect_WT['압연월'].fillna("", inplace=True)  # Fill missing values with "" (or any other appropriate value)

# 압연월은 년월 컬럼 앞에 배치
position = Defect_WT.columns.get_loc('년월일')
Defect_WT.insert(position, '압연월', Defect_WT.pop('압연월'))
Defect_WT.head()
```

```python
#압연년 구분
Defect_WT['압연년'] = Defect_WT['최종압연일자'].dt.year
#Defect_WT['압연년'] = Defect_WT['압연년'].fillna("", inplace=True)  # Fill missing values with "" (or any other appropriate value)

# 압연년은 압연월 컬럼 앞에 배치
position = Defect_WT.columns.get_loc('압연월')
Defect_WT.insert(position, '압연년', Defect_WT.pop('압연년'))
Defect_WT.head()
```

```python
#압연 년월을 압연 Roll 단위 구분
Defect_WT['압연일'] = Defect_WT['최종압연일자'].dt.day
#efect_WT['압연일'] = Defect_WT['압연일'].fillna("", inplace=True)  # Fill missing values with "" (or any other appropriate value)
character = "-"

def Roll_Mth(row):
    if pd.notna(row['압연일']):
        if row['압연일'] > 10:
            return row['압연월']
        else:
            return row['압연월'] - 1
    else:
        return ""
    
Defect_WT['Roll_Mth'] = Defect_WT.apply(Roll_Mth, axis=1)

# Roll_Mth은 R_Year 컬럼 앞에 배치
position = Defect_WT.columns.get_loc('압연년')
Defect_WT.insert(position, 'Roll_Mth', Defect_WT.pop('Roll_Mth'))

Defect_WT['Roll_Yr_Mth'] = Defect_WT['압연년'].astype(str) + character + Defect_WT['Roll_Mth'].astype(str)

# Check if '압연년' or 'Roll_Mth' is empty and replace 'Roll_Yr_Mth' with a blank space
#mask = (Defect_WT['압연년'] == "") | (Defect_WT['Roll_Mth'] == "")
#Defect_WT.loc[mask, 'Roll_Yr_Mth'] = ""

# Roll_Yr_Mth은 Roll_Mth 컬럼 앞에 배치
position = Defect_WT.columns.get_loc('Roll_Mth')
Defect_WT.insert(position, 'Roll_Yr_Mth', Defect_WT.pop('Roll_Yr_Mth'))

Defect_WT.head()
#print(Defect_WT['Roll_Yr_Mth'])
```

```python
# Check if 'Roll_Mth' is empty and replace 'Roll_분기' with a blank space
mask = Defect_WT['Roll_Mth'] == ""
Defect_WT.loc[mask, 'Roll_분기'] = ""

# Calculate 'Roll_분기' for rows with non-empty 'Roll_Mth'
mask = Defect_WT['Roll_Mth'] != ""
Defect_WT.loc[mask, 'Roll_분기'] = Defect_WT.loc[mask, 'Roll_Mth'].apply(lambda month: '1Q' if month <= 3 else '2Q' if month <= 6 else '3Q' if month <= 9 else '4Q')

# R_Quarter은 Roll_Yr_Mth 컬럼 앞에 배치
position = Defect_WT.columns.get_loc('Roll_Yr_Mth')
Defect_WT.insert(position, 'Roll_분기', Defect_WT.pop('Roll_분기'))

Defect_WT.head()
#print(Defect_WT['R_Quarter'])
```

```python
# 1,433처럼 중량에 콤마가 있는 Object는 콤마를 제거 후 숫자로 변환해야 함.
Defect_WT['발생중량'] = Defect_WT['발생중량'].str.replace(',', '').astype(int)
#Data_type = Defect_df['발생중량'].dtypes
#print(Data_type)

Defect_WT.head()
```
